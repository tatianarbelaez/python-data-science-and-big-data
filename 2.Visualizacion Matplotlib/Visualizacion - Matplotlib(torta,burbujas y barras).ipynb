```python
plt.pie(x =data[1],
       labels = data[0])
plt.show()
#Con este codigo se crea grafico de torta y se muestra
```

```python
import seaborn as sns
np.random.seed(0)
#Con este comando si se vuelve a correr el codigo da lo mismo
df = df.sample(frac=1).head(100)
```

```python
plt.scatter(x=df['High CI'],y=df['Low CI'],s=df['Value'])
#Genera una grafica de Burbujas 
#Solo se hace cuando todas las variables son numericas 
```

```python
plt.scatter(x=df['High CI'],y=df['Low CI'],s=df['Value'],alpha =.3,c= df['Value'].isin([9:12]))
#Para hacer las burbujas opacas alpha = 1, para burbujas invisibles alpha = 0 
#luego, se puede hacer un filtro por el nombre de pais, usando una lista
#si antes del filtro agrego c= entonces pone diferentes colores a las burbujas segun nombre del pais 
plt.title("High and low CI values by age and sex")
#Le pone titulo al grafico
plt.xlabel(s='High CI')
#le pone titulo al eje x
plt.ylabel(s='Low CI')
#le pone titulo al eje y
plt.ylim([0,150])
#en que rango de y quiero visualizar los datos 
plt.xticks([2,4,5])
#en que rango o en cuales observaciones de x quiero visualizar los datos
plt.text(x=28,y=37,s='yo')
#Si quiero elegir unos datos especificos en el grafico y ponerles un nombre
```

```python
plt.scatter(x=df['High CI'],y=df['Low CI'],s=df['Value'],alpha =.3, c= df['Value'].isin([12.6,17.7]))
plt.title("High and low CI values")
plt.xlabel('High CI')
plt.ylabel('Low CI')
plt.text(x=13.3,y=12.1,s='yo')
```

```python
plt.scatter(x=df['High CI'],y=df['Low CI'],s=df['Value'],alpha =.4, c= df['Value'] > 30)
plt.title("High and low CI values")
plt.xlabel('High CI')
plt.ylabel('Low CI')
plt.text(x=13.3,y=12.1,s='yo')
plt.ylim([20,80])
```

```python
import seaborn as sns
np.random.seed(0)
plt.pie(x =data[1],
       labels = data[0],
       colors = sns.color_palette('hls',7),
       explode = (0,0.2,0,0,0,0,0,0,0.1),
       startangle = 90,
       radius = 1,
       autopct = '%1.1f%%',
       labeldistance = 1)
plt.legend(loc='lower right')
plt.show()
#el paquete seaborn tiene diferentes paletas de colores que pueden utilizarse en los graficos
#la funcion explode resalta la porcion del grafico de torta que quieras
#la funcion autopct hace que cada porcion diga el porcentaje que representa del total 
#la funcion labeldistance indica que tan cerca o lejos queda la etiqueta del grafico
#la funcion legend crea leyendas y con loc se puede decidir donde debe ir, si se pone un lugar no permitido sale error y python
#da las posibles opciones
```
