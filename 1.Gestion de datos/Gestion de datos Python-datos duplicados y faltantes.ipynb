```python
# -*- coding: utf-8 -*-
import pandas as pd
df_finance = pd.read_csv(r"C:\Users\user\Downloads\catalogo_cf.csv", encoding='iso-8859-1', nrows = 1000000)
#nrows=1e6 un millon de filas 
```

```python
df_duplicate = df_finance.append(df_finance)
```

```python
df_duplicate = df_duplicate.sample(frac=1)
#la base duplicada es solo una muestra al azar de la base 
```

```python
df_clean = df_duplicate.drop_duplicates()
#elimina todos los casos duplicados
```

```python
len(df_clean) == len(df_finance)
```

```python
df_clean.drop_duplicates(subset = 'Nombre de la serie ')
#Elimina los duplicados que hay en nombre de la serie
```
