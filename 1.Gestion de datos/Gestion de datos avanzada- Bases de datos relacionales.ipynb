```python
################################################################################################################################
# Bases de datos relacionales JOINS
```

```python
clientes = [("A","MOVIL"),("B","MOVIL"),("A","PORTATIL"),("A","TABLET"),("B","TABLET"),("C","PORTATIL"),("D","SMARTWATCH"),
            ("E","CONSOLA")]
cli_labels = ["cliente","producto"]

cli_df = pd.DataFrame.from_records(clientes,columns = cli_labels)
```

```python
productores = [("a","MOVIL"),("a","SMARTWATCH"),("a","TABLET"),("b","PORTATIL"),("c","SOBREMESA"),("c","PORTATIL")]
prod_labels = ["productor","producto"]

prod_df = pd.DataFrame.from_records(productores,columns = prod_labels)
```

```python
cli_df
```

```python
prod_df
```

```python
pd.merge(cli_df,prod_df,on="producto",how="outer")
#on = la llave que aparece en ambas tablas
#Si la llave no se encuentra en una de las dos tablas, no se llena la informacion, CONSOLA no se encuentra en la tabla de productores
#mientras que SOBREMESA no esta en la de clientes
```

```python
pd.merge(cli_df,prod_df,on="producto",how="inner")
#La diferencia con el outer, es que elimina las filas en las que no se encontró la llave en ambas tablas
```

```python
prod_df
```

```python
pd.merge(cli_df,prod_df,on="producto",how="right")
#Indicamos que junte el dataframe de la izquierda con el de la derecha solo si sus valores aparecen en el de la derecha
#vemos que aprece todo lo que habia en el dataframe de la derecha prod_df y se le unió la informacion del otro (solo la que coincide)
```

```python
pd.merge(cli_df,prod_df,on="producto",how="left")
#Sucede lo contrario al caso anterior, busca todas las claves de producto que habia en cli_df y le une solo los que coinciden de prod_df
```
