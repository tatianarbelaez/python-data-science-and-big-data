```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\user\Documents\PYTHON\Python data science y big data\Datos\Indicators_of_Health_Insurance_Coverage.csv",nrows=1e5)
newdf = df[['Low CI','High CI']].dropna()
```

```python
newdf = df[['Low CI','High CI','Value','Time Period']].dropna()
```

```python
kmeans = KMeans(n_clusters=3,random_state=0,n_jobs=-1).fit(newdf)
#se le indican el numero de clusters, random_state es para que al replicar kmeans de igual, n_jobs hace que usemos todos los 
#procesadores del computador para acelerar el proceso
kmeans.labels_
#El arreglo que obtenemos nos dice cada fila a que cluster pertenece
```

```python
np.unique(kmeans.labels_,return_counts=True)
#muestra como quedaron distribuidas las oservaciones de cada cluster
```

```python
plt.scatter(newdf['Low CI'],newdf['High CI'],c=kmeans.labels_)
plt.show()
```

```python
kmeans.cluster_centers_
#muestra alrededor de que centros, de que punto en el plano, se agrupa cada cluster
```
